```python
import pandas as pd
import numpy as np
```

```python
rec1 = pd.read_csv('/kaggle/input/dataset/firstSub.csv')
rec1
```

```python
rec2 = pd.read_csv('/kaggle/input/dataset/secondSub.csv')
rec2
```

```python
%load_ext Cython
```

```python
TOTAL_RECOMMENDATIONS = []
recommendations = []
for user_id in range(0,41116):
    rec1_recommendations = list(map(int, rec1["item_list"][user_id].split()))
    rec2_recommendations = list(map(int, rec2["item_list"][user_id].split()))
    for i in range(0,10):
        recommendations.append(rec2_recommendations[i])
        recommendations.append(rec1_recommendations[i])

    recommendations = list(dict.fromkeys(recommendations))
    TOTAL_RECOMMENDATIONS.append(recommendations[0:10])
    recommendations.clear()
    
```

```python
len(TOTAL_RECOMMENDATIONS)
```

```python
test_users = pd.read_csv('/kaggle/input/competition-data/data_target_users_test.csv')
test_users
```

```python
for index in range(len(TOTAL_RECOMMENDATIONS)):
    TOTAL_RECOMMENDATIONS[index]=np.array(TOTAL_RECOMMENDATIONS[index])
    
test_users['item_list']= TOTAL_RECOMMENDATIONS
test_users['item_list'] = pd.DataFrame([str(line).strip('[').strip(']').replace("'","") for line in test_users['item_list']])
test_users.to_csv('recommendations.csv', index=False)
```
